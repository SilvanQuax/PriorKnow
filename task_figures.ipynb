```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 16})

X = np.linspace(-20,20,1000)

Y1  = np.exp(-(X-5)**2/(2*25))
Y2  = np.exp(-(X+5)**2/(2*25))

Y3  = np.exp(-(X)**2/(2*5))/np.sqrt(2*np.pi*5)
Y4  = np.exp(-(X)**2/(2*10))/np.sqrt(2*np.pi*10)
Y5  = np.exp(-(X)**2/(2*25))/np.sqrt(2*np.pi*25)
Y6  = np.exp(-(X)**2/(2*50))/np.sqrt(2*np.pi*50)

Y7  = np.exp(-(X)**2/(2*100))/np.sqrt(2*np.pi*100)


plt.figure(figsize = [4,3.3])
plt.plot(X,Y1,'--k')
plt.plot(X,Y2,'k')

plt.yticks([])
plt.xlabel('Stimulus')
plt.ylabel('Probability')
plt.ylim([0,1.3])
plt.tight_layout()
#plt.savefig('/vol/ccnlab1/silqua/prirep/figures/pytorch/class_task.eps',format='eps')

plt.figure(figsize = [4,3.3])
plt.plot(X,Y3,color = [0,0,0])
plt.plot(X,Y4,color = [0.2,0.2,0.2])
plt.plot(X,Y5,color = [0.4,0.4,0.4])
plt.plot(X,Y6,color = [0.6,0.6,0.6])
plt.plot(X,Y7,color = [0.8,0.8,0.8])


plt.yticks([])
plt.xlabel('Stimulus')
plt.ylabel('Probability')
plt.ylim([0,0.23])
plt.tight_layout()
#plt.savefig('/vol/ccnlab1/silqua/prirep/figures/pytorch/cont_task.eps',format='eps')
```
